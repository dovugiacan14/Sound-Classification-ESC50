```python
# ‚úÖ Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import os
import numpy as np
import pandas as pd
import librosa
import soundfile as sf
import random
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# ‚úÖ C·∫•u h√¨nh seed
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ‚úÖ ƒê·ªãnh nghƒ©a ƒë∆∞·ªùng d·∫´n d·ªØ li·ªáu
DATA_PATH = "ESC-50-master/"
CSV_PATH = os.path.join(DATA_PATH, "meta/esc50.csv")
AUDIO_PATH = os.path.join(DATA_PATH, "audio")
AUGMENTED_PATH = "ESC-50-augmented-data"

# ‚úÖ ƒê·ªçc d·ªØ li·ªáu g·ªëc
dataset = pd.read_csv(CSV_PATH)
dataset["files_path"] = dataset["filename"].apply(lambda x: os.path.join(AUDIO_PATH, x))

# ‚úÖ M√£ h√≥a nh√£n
encoder = LabelEncoder()
dataset["target"] = encoder.fit_transform(dataset["category"])
num_classes = len(encoder.classes_)

# ‚úÖ Tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng √¢m thanh
features = ['zero_crossing_rate', 'chroma_stft', 'rmse', 'spectral_centroid', 'spectral_bandwidth',
            'beat_per_minute', 'rolloff'] + [f'mfcc_{i}' for i in range(20)]

for feature in features:
    dataset[feature] = np.zeros(len(dataset))

for i in tqdm(range(len(dataset))):
    file_path = dataset['files_path'][i]
    if os.path.exists(file_path):
        y, sr = librosa.load(file_path, mono=True)
        dataset.at[i, 'rmse'] = np.mean(librosa.feature.rms(y=y))
        dataset.at[i, 'chroma_stft'] = np.mean(librosa.feature.chroma_stft(y=y, sr=sr))
        dataset.at[i, 'spectral_centroid'] = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
        dataset.at[i, 'spectral_bandwidth'] = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
        dataset.at[i, 'rolloff'] = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
        dataset.at[i, 'zero_crossing_rate'] = np.mean(librosa.feature.zero_crossing_rate(y))
        tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
        dataset.at[i, 'beat_per_minute'] = tempo
        mfcc = librosa.feature.mfcc(y=y, sr=sr)
        for e in range(20):
            dataset.at[i, f'mfcc_{e}'] = np.mean(mfcc[e])

# ‚úÖ Chia t·∫≠p d·ªØ li·ªáu train/test
train = dataset.sample(frac=0.8, random_state=SEED)
test = dataset.drop(train.index)

X_train = train[features]
X_test = test[features]
y_train = train['target']
y_test = test['target']

scaler = MinMaxScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

# ‚úÖ X√¢y d·ª±ng 3 m√¥ h√¨nh CNN, LSTM, Transformer
def build_cnn(input_shape, num_classes):
    model = models.Sequential([
        layers.Dense(512, activation="relu", input_shape=input_shape),
        layers.BatchNormalization(),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation="softmax")
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model

def attention_layer(inputs):
    """
    C∆° ch·∫ø Attention gi√∫p m√¥ h√¨nh t·∫≠p trung v√†o ƒë·∫∑c tr∆∞ng quan tr·ªçng.
    """
    attention_weights = layers.Dense(1, activation="tanh")(inputs)
    attention_weights = layers.Flatten()(attention_weights)
    attention_weights = layers.Activation("softmax")(attention_weights)
    attention_weights = layers.RepeatVector(inputs.shape[-1])(attention_weights)
    attention_weights = layers.Permute([2, 1])(attention_weights)
    
    return layers.Multiply()([inputs, attention_weights])

def build_lstm(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)
    
    # ‚úÖ Th√™m CNN ƒë·ªÉ tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng tr∆∞·ªõc khi ƒë∆∞a v√†o LSTM
    x = layers.Reshape((input_shape[0], 1))(inputs)
    x = layers.Conv1D(128, kernel_size=3, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    
    # ‚úÖ Bi-directional LSTM ƒë·ªÉ t·∫≠n d·ª•ng ng·ªØ c·∫£nh hai chi·ªÅu
    x = layers.Bidirectional(layers.LSTM(128, return_sequences=True))(x)
    x = layers.Bidirectional(layers.LSTM(64, return_sequences=True))(x)

    # ‚úÖ Th√™m Attention
    x = attention_layer(x)

    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),  # Gi·∫£m learning rate
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model


def build_transformer(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # ‚úÖ Chuy·ªÉn ƒë·ªïi input t·ª´ 2D -> 3D (sequence_length = s·ªë ƒë·∫∑c tr∆∞ng, channels = 1)
    x = layers.Reshape((input_shape[0], 1))(inputs)  

    # ‚úÖ Positional Encoding (th√™m th√¥ng tin v·ªÅ v·ªã tr√≠)
    x = layers.Conv1D(filters=128, kernel_size=3, padding="same", activation="relu")(x)
    x = layers.LayerNormalization()(x)

    # ‚úÖ Multi-Head Attention v·ªõi nhi·ªÅu t·∫ßng
    for _ in range(3):  # L·∫∑p 3 l·∫ßn ƒë·ªÉ tƒÉng ƒë·ªô s√¢u c·ªßa m√¥ h√¨nh
        attn_output = layers.MultiHeadAttention(num_heads=8, key_dim=64)(x, x)
        x = layers.Add()([x, attn_output])
        x = layers.LayerNormalization()(x)
    
    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),  # Gi·∫£m learning rate
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model



# ‚úÖ Hu·∫•n luy·ªán 3 m√¥ h√¨nh
models_dict = {
    "CNN": build_cnn((X_train.shape[1],), num_classes),
    "LSTM": build_lstm((X_train.shape[1],), num_classes),
    "Transformer": build_transformer((X_train.shape[1],), num_classes)
}

results = []
for name, model in models_dict.items():
    print(f"\nüîµ Training {name}...")
    history =model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test), verbose=1)
    train_acc = history.history['accuracy'][-1]  # Get the last epoch's training accuracy
    train_loss = history.history['loss'][-1]      # Get the last epoch's training loss
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    results.append({
            "Model": name,
            "Train Accuracy": train_acc,
            "Train Loss": train_loss,
            "Test Accuracy": test_acc,
            "Test Loss": test_loss
        })

# ‚úÖ B·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£
df_results = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)
df_results_styled = df_results.style.background_gradient(cmap="coolwarm")
print(df_results_styled)
sns.barplot(x="Test Accuracy", y="Model", data=df_results, palette="coolwarm")
plt.xlabel("Accuracy (%)")
plt.ylabel("Model")
plt.title("So s√°nh ƒë·ªô ch√≠nh x√°c c·ªßa c√°c m√¥ h√¨nh")
plt.show()

# ‚úÖ S·∫Øp x·∫øp top model theo ƒë·ªô ch√≠nh x√°c
df_summary = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)


# ‚úÖ Hi·ªÉn th·ªã b·∫£ng x·∫øp h·∫°ng top model
print("\nüîπ B·∫£ng X·∫øp H·∫°ng Top Models:")
print(df_summary.to_string(index=False))
# ‚úÖ V·∫Ω bi·ªÉu ƒë·ªì Top Models
plt.figure(figsize=(10, 6))
sns.barplot(x="Test Accuracy", y="Model", data=df_summary, palette="coolwarm")
plt.xlabel("Test Accuracy (%)")
plt.ylabel("Model")
plt.title("Top Model Performance")
plt.show()
# ‚úÖ L·∫•y 1000 m·∫´u ng·∫´u nhi√™n t·ª´ t·∫≠p test g·ªëc
random_indices = np.random.choice(len(X_test), 400, replace=False)

# ‚úÖ D·ª± ƒëo√°n nh√£n cho t·ª´ng m√¥ h√¨nh
model_results = []
summary_stats = []

for name, model in models_dict.items():
    print(f"\nüü¢ ƒêang d·ª± ƒëo√°n v·ªõi m√¥ h√¨nh: {name}")
    y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
    y_true = y_test.iloc[random_indices].values

    # ‚úÖ Chuy·ªÉn nh√£n v·ªÅ d·∫°ng t√™n th·ª±c t·∫ø
    true_labels = encoder.inverse_transform(y_true)
    predicted_labels = encoder.inverse_transform(y_pred)

    # ‚úÖ T·ªïng h·ª£p k·∫øt qu·∫£ d·ª± ƒëo√°n
    results = []
    for i in range(len(random_indices)):
        results.append({
            "Model": name,
            "File": dataset.iloc[random_indices[i]]["files_path"],
            "True Label": true_labels[i],
            "Predicted Label": predicted_labels[i],
            "Correct": "‚úÖ" if true_labels[i] == predicted_labels[i] else "‚ùå"
        })

    df_results = pd.DataFrame(results)

    # ‚úÖ T√≠nh to√°n s·ªë d·ª± ƒëo√°n ƒë√∫ng/sai
    correct_predictions = sum(y_true == y_pred)
    incorrect_predictions = len(y_true) - correct_predictions
    accuracy = round((correct_predictions / len(y_true)) * 100, 2)

    # ‚úÖ Xu·∫•t b·∫£ng t·ªïng h·ª£p
    summary_stats.append({
        "Model": name,
        "Correct Predictions": correct_predictions,
        "Incorrect Predictions": incorrect_predictions,
        "Accuracy (%)": accuracy
    })

    # ‚úÖ L∆∞u k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
    model_results.append(df_results)

# ‚úÖ Hi·ªÉn th·ªã b·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£ Accuracy
df_summary = pd.DataFrame(summary_stats)

print("\nüîπ B·∫£ng T·ªïng H·ª£p Accuracy:")
print(df_summary.to_string(index=False))

# ‚úÖ Hi·ªÉn th·ªã b·∫£ng chi ti·∫øt k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
for df in model_results:
    model_name = df["Model"][0]
    print(f"\nüîπ B·∫£ng Chi Ti·∫øt D·ª± ƒêo√°n - {model_name}:")
    print(df[['File', 'True Label', 'Predicted Label', 'Correct']].head(10).to_string(index=False))  # Hi·ªÉn th·ªã 10 d√≤ng ƒë·∫ßu ti√™n

# ‚úÖ Hi·ªÉn th·ªã ma tr·∫≠n nh·∫ßm l·∫´n cho t·ª´ng m√¥ h√¨nh
plt.figure(figsize=(15, 10))
for i, (name, model) in enumerate(models_dict.items()):
    y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
    y_true = y_test.iloc[random_indices].values

    plt.subplot(2, 3, i + 1)  # ƒêi·ªÅu ch·ªânh ƒë·ªÉ ph√π h·ª£p v·ªõi s·ªë m√¥ h√¨nh
    sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt="d", cmap="coolwarm")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {name}")

plt.tight_layout()
plt.show()

# ‚úÖ Xu·∫•t b√°o c√°o ph√¢n lo·∫°i cho t·ª´ng m√¥ h√¨nh
for name, model in models_dict.items():
    y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
    y_true = y_test.iloc[random_indices].values

    print(f"\nüîπ Classification Report - {name}:")
    print(classification_report(y_true, y_pred, labels=np.unique(y_true), target_names=encoder.inverse_transform(np.unique(y_true))))


```

```python
# ‚úÖ Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import os
import numpy as np
import pandas as pd
import librosa
import soundfile as sf
import random
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ‚úÖ C·∫•u h√¨nh seed
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ‚úÖ ƒê·ªãnh nghƒ©a ƒë∆∞·ªùng d·∫´n d·ªØ li·ªáu
DATA_PATH = "ESC-50-master/"
CSV_PATH = os.path.join(DATA_PATH, "meta/esc50.csv")
AUDIO_PATH = os.path.join(DATA_PATH, "audio")
AUGMENTED_PATH = "ESC-50-augmented-data"

# ‚úÖ ƒê·ªçc d·ªØ li·ªáu g·ªëc
dataset = pd.read_csv(CSV_PATH)
dataset["files_path"] = dataset["filename"].apply(lambda x: os.path.join(AUDIO_PATH, x))

# ‚úÖ M√£ h√≥a nh√£n
encoder = LabelEncoder()
dataset["target"] = encoder.fit_transform(dataset["category"])
num_classes = len(encoder.classes_)

# ‚úÖ T·∫°o th∆∞ m·ª•c tƒÉng c∆∞·ªùng
os.makedirs(AUGMENTED_PATH, exist_ok=True)

# ‚úÖ H√†m l∆∞u √¢m thanh tƒÉng c∆∞·ªùng
def save_augmented_audio(filename, data, sr=16000):
    sf.write(filename, data, sr)

# ‚úÖ C√°c ph∆∞∆°ng ph√°p tƒÉng c∆∞·ªùng d·ªØ li·ªáu
def add_noise(data):
    return data + np.random.normal(0, 0.002, len(data))

def pitch_shifting(data, sr=16000):
    return librosa.effects.pitch_shift(data.astype(np.float32), sr=sr, n_steps=np.random.uniform(-2, 2))

def time_stretching(data, rate=1.5):
    return librosa.effects.time_stretch(data.astype(np.float32), rate=rate)

def volume_change(data, gain=3):
    return data * (10**(gain / 20))

# ‚úÖ Th·ª±c hi·ªán tƒÉng c∆∞·ªùng d·ªØ li·ªáu
print("\nüü¢ B·∫Øt ƒë·∫ßu th·ª±c hi·ªán Data Augmentation...")
augmented_files = []
augmented_targets = []

for index, row in tqdm(dataset.iterrows(), total=len(dataset)):
    file_path = row["files_path"]
    label = row["target"]
    
    try:
        data, sr = librosa.load(file_path, sr=16000)
        augmentations = [add_noise(data), pitch_shifting(data, sr=sr), time_stretching(data), volume_change(data)]

        for idx, aug_data in enumerate(augmentations):
            aug_filename = f"{row['filename'][:-4]}_aug_{idx}.wav"
            aug_filepath = os.path.join(AUGMENTED_PATH, aug_filename)
            save_augmented_audio(aug_filepath, aug_data, sr)
            
            augmented_files.append(aug_filepath)
            augmented_targets.append(label)
    
    except Exception as e:
        print(f"‚ö†Ô∏è L·ªói khi tƒÉng c∆∞·ªùng d·ªØ li·ªáu cho file {file_path}: {e}")

# ‚úÖ T·∫°o DataFrame d·ªØ li·ªáu tƒÉng c∆∞·ªùng v√† h·ª£p nh·∫•t v·ªõi t·∫≠p g·ªëc
dataset_aug = pd.DataFrame({"files_path": augmented_files, "target": augmented_targets})
dataset = pd.concat([dataset, dataset_aug], ignore_index=True)

print(f"\nüü¢ T·ªïng s·ªë m·∫´u sau tƒÉng c∆∞·ªùng: {len(dataset)}")

# ‚úÖ Tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng √¢m thanh
features = ['zero_crossing_rate', 'chroma_stft', 'rmse', 'spectral_centroid', 'spectral_bandwidth',
            'beat_per_minute', 'rolloff'] + [f'mfcc_{i}' for i in range(20)]

for feature in features:
    dataset[feature] = np.zeros(len(dataset))

print("\nüü¢ B·∫Øt ƒë·∫ßu tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng...")
for i in tqdm(range(len(dataset))):
    file_path = dataset['files_path'][i]
    if os.path.exists(file_path):
        y, sr = librosa.load(file_path, mono=True)
        dataset.at[i, 'rmse'] = np.mean(librosa.feature.rms(y=y))
        dataset.at[i, 'chroma_stft'] = np.mean(librosa.feature.chroma_stft(y=y, sr=sr))
        dataset.at[i, 'spectral_centroid'] = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
        dataset.at[i, 'spectral_bandwidth'] = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
        dataset.at[i, 'rolloff'] = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
        dataset.at[i, 'zero_crossing_rate'] = np.mean(librosa.feature.zero_crossing_rate(y))
        tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
        dataset.at[i, 'beat_per_minute'] = tempo
        mfcc = librosa.feature.mfcc(y=y, sr=sr)
        for e in range(20):
            dataset.at[i, f'mfcc_{e}'] = np.mean(mfcc[e])

# ‚úÖ Chia t·∫≠p d·ªØ li·ªáu train/test
train = dataset.sample(frac=0.8, random_state=SEED)
test = dataset.drop(train.index)

X_train = train[features]
X_test = test[features]
y_train = train['target']
y_test = test['target']

scaler = MinMaxScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

# ‚úÖ X√¢y d·ª±ng m√¥ h√¨nh CNN
def build_cnn(input_shape, num_classes):
    model = models.Sequential([
        layers.Dense(512, activation="relu", input_shape=input_shape),
        layers.BatchNormalization(),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation="softmax")
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
def attention_layer(inputs):
    """
    C∆° ch·∫ø Attention gi√∫p m√¥ h√¨nh t·∫≠p trung v√†o ƒë·∫∑c tr∆∞ng quan tr·ªçng.
    """
    attention_weights = layers.Dense(64, activation="tanh")(inputs)
    attention_weights = layers.Dense(1, activation="softmax")(attention_weights)
    attention_weights = layers.Multiply()([inputs, attention_weights])

    return layers.GlobalAveragePooling1D()(attention_weights)
    
   

def build_lstm(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # ‚úÖ Th√™m t·∫ßng CNN ƒë·ªÉ tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng
    x = layers.Reshape((input_shape[0], 1))(inputs)
    x = layers.Conv1D(256, kernel_size=5, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(128, kernel_size=3, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)

    # ‚úÖ Bi-directional LSTM ƒë·ªÉ t·∫≠n d·ª•ng ng·ªØ c·∫£nh hai chi·ªÅu
    x = layers.Bidirectional(layers.LSTM(256, return_sequences=True, dropout=0.3))(x)
    x = layers.Bidirectional(layers.LSTM(128, return_sequences=True, dropout=0.3))(x)
    x = layers.Bidirectional(layers.LSTM(64, return_sequences=True, dropout=0.3))(x)

    # ‚úÖ Th√™m Attention ƒë·ªÉ t·∫≠p trung v√†o ƒë·∫∑c tr∆∞ng quan tr·ªçng
    x = attention_layer(x)

    # ‚úÖ Th√™m BatchNormalization ƒë·ªÉ ·ªïn ƒë·ªãnh hu·∫•n luy·ªán
    x = layers.BatchNormalization()(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003), 
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    
    return model


def build_transformer(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # ‚úÖ Chuy·ªÉn ƒë·ªïi input t·ª´ 2D -> 3D (sequence_length = s·ªë ƒë·∫∑c tr∆∞ng, channels = 1)
    x = layers.Reshape((input_shape[0], 1))(inputs)  

    # ‚úÖ Positional Encoding (th√™m th√¥ng tin v·ªÅ v·ªã tr√≠)
    x = layers.Conv1D(filters=128, kernel_size=3, padding="same", activation="relu")(x)
    x = layers.LayerNormalization()(x)

    # ‚úÖ Multi-Head Attention v·ªõi nhi·ªÅu t·∫ßng
    for _ in range(3):  # L·∫∑p 3 l·∫ßn ƒë·ªÉ tƒÉng ƒë·ªô s√¢u c·ªßa m√¥ h√¨nh
        attn_output = layers.MultiHeadAttention(num_heads=16, key_dim=64)(x, x)
        x = layers.Add()([x, attn_output])
        x = layers.LayerNormalization()(x)
    
    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),  # Gi·∫£m learning rate
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model


# ‚úÖ Hu·∫•n luy·ªán m√¥ h√¨nh
models_dict = {
    "CNN": build_cnn((X_train.shape[1],), num_classes),
    "LSTM": build_lstm((X_train.shape[1],), num_classes),
    "Transformer": build_transformer((X_train.shape[1],), num_classes)
}

results = []
for name, model in models_dict.items():
    print(f"\nüîµ Training {name}...")
    history =model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test), verbose=1)
    train_acc = history.history['accuracy'][-1]  # Get the last epoch's training accuracy
    train_loss = history.history['loss'][-1]      # Get the last epoch's training loss
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    results.append({
            "Model": name,
            "Train Accuracy": train_acc,
            "Train Loss": train_loss,
            "Test Accuracy": test_acc,
            "Test Loss": test_loss
        })

# ‚úÖ B·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£
df_results = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)
df_results_styled = df_results.style.background_gradient(cmap="coolwarm")
print(df_results_styled)
sns.barplot(x="Test Accuracy", y="Model", data=df_results, palette="coolwarm")
plt.xlabel("Accuracy (%)")
plt.ylabel("Model")
plt.title("So s√°nh ƒë·ªô ch√≠nh x√°c c·ªßa c√°c m√¥ h√¨nh")
plt.show()

# ‚úÖ S·∫Øp x·∫øp top model theo ƒë·ªô ch√≠nh x√°c
df_summary = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)


# ‚úÖ Hi·ªÉn th·ªã b·∫£ng x·∫øp h·∫°ng top model
print("\nüîπ B·∫£ng X·∫øp H·∫°ng Top Models:")
print(df_summary.to_string(index=False))

# ‚úÖ V·∫Ω bi·ªÉu ƒë·ªì Top Models
plt.figure(figsize=(10, 6))
sns.barplot(x="Test Accuracy", y="Model", data=df_summary, palette="coolwarm")
plt.xlabel("Test Accuracy (%)")
plt.ylabel("Model")
plt.title("Top Model Performance")
plt.show()
# ‚úÖ L·∫•y 1000 m·∫´u ng·∫´u nhi√™n t·ª´ t·∫≠p test g·ªëc
random_indices = np.random.choice(len(X_test), 1000, replace=False)

# ‚úÖ D·ª± ƒëo√°n nh√£n cho t·ª´ng m√¥ h√¨nh
model_results = []
summary_stats = []

for name, model in models_dict.items():
    print(f"\nüü¢ ƒêang d·ª± ƒëo√°n v·ªõi m√¥ h√¨nh: {name}")
    y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
    y_true = y_test.iloc[random_indices].values

    # ‚úÖ Chuy·ªÉn nh√£n v·ªÅ d·∫°ng t√™n th·ª±c t·∫ø
    true_labels = encoder.inverse_transform(y_true)
    predicted_labels = encoder.inverse_transform(y_pred)

    # ‚úÖ T·ªïng h·ª£p k·∫øt qu·∫£ d·ª± ƒëo√°n
    results = []
    for i in range(len(random_indices)):
        results.append({
            "Model": name,
            "File": dataset.iloc[random_indices[i]]["files_path"],
            "True Label": true_labels[i],
            "Predicted Label": predicted_labels[i],
            "Correct": "‚úÖ" if true_labels[i] == predicted_labels[i] else "‚ùå"
        })

    df_results = pd.DataFrame(results)

    # ‚úÖ T√≠nh to√°n s·ªë d·ª± ƒëo√°n ƒë√∫ng/sai
    correct_predictions = sum(y_true == y_pred)
    incorrect_predictions = len(y_true) - correct_predictions
    accuracy = round((correct_predictions / len(y_true)) * 100, 2)

    # ‚úÖ Xu·∫•t b·∫£ng t·ªïng h·ª£p
    summary_stats.append({
        "Model": name,
        "Correct Predictions": correct_predictions,
        "Incorrect Predictions": incorrect_predictions,
        "Accuracy (%)": accuracy
    })

    # ‚úÖ L∆∞u k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
    model_results.append(df_results)

# ‚úÖ Hi·ªÉn th·ªã b·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£ Accuracy
df_summary = pd.DataFrame(summary_stats)

print("\nüîπ B·∫£ng T·ªïng H·ª£p Accuracy:")
print(df_summary.to_string(index=False))

# ‚úÖ Hi·ªÉn th·ªã b·∫£ng chi ti·∫øt k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
for df in model_results:
    model_name = df["Model"][0]
    print(f"\nüîπ B·∫£ng Chi Ti·∫øt D·ª± ƒêo√°n - {model_name}:")
    print(df[['File', 'True Label', 'Predicted Label', 'Correct']].head(10).to_string(index=False))  # Hi·ªÉn th·ªã 10 d√≤ng ƒë·∫ßu ti√™n

# ‚úÖ Hi·ªÉn th·ªã ma tr·∫≠n nh·∫ßm l·∫´n cho t·ª´ng m√¥ h√¨nh
plt.figure(figsize=(15, 10))
for i, (name, model) in enumerate(models_dict.items()):
    y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
    y_true = y_test.iloc[random_indices].values

    plt.subplot(2, 3, i + 1)  # ƒêi·ªÅu ch·ªânh ƒë·ªÉ ph√π h·ª£p v·ªõi s·ªë m√¥ h√¨nh
    sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt="d", cmap="coolwarm")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title(f"Confusion Matrix - {name}")

plt.tight_layout()
plt.show()

# ‚úÖ Xu·∫•t b√°o c√°o ph√¢n lo·∫°i cho t·ª´ng m√¥ h√¨nh
for name, model in models_dict.items():
    y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
    y_true = y_test.iloc[random_indices].values

    print(f"\nüîπ Classification Report - {name}:")
    print(classification_report(y_true, y_pred, labels=np.unique(y_true), target_names=encoder.inverse_transform(np.unique(y_true))))
```

```python
# ‚úÖ Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import os
import numpy as np
import pandas as pd
import librosa
import soundfile as sf
import random
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# ‚úÖ C·∫•u h√¨nh seed
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
# ‚úÖ ƒê·ªãnh nghƒ©a ƒë∆∞·ªùng d·∫´n d·ªØ li·ªáu
DATA_PATH = "ESC-50-master/"
CSV_PATH = os.path.join(DATA_PATH, "meta/esc50.csv")
AUDIO_PATH = os.path.join(DATA_PATH, "audio")
AUGMENTED_PATH = "ESC-50-augmented-data"
# ‚úÖ ƒê·ªçc d·ªØ li·ªáu g·ªëc
dataset = pd.read_csv(CSV_PATH)
dataset["files_path"] = dataset["filename"].apply(lambda x: os.path.join(AUDIO_PATH, x))
# ‚úÖ M√£ h√≥a nh√£n
encoder = LabelEncoder()
dataset["target"] = encoder.fit_transform(dataset["category"])
num_classes = len(encoder.classes_)
# ‚úÖ T·∫°o th∆∞ m·ª•c tƒÉng c∆∞·ªùng
os.makedirs(AUGMENTED_PATH, exist_ok=True)
# ‚úÖ H√†m l∆∞u √¢m thanh tƒÉng c∆∞·ªùng
def save_augmented_audio(filename, data, sr=16000):
   sf.write(filename, data, sr)
# ‚úÖ C√°c ph∆∞∆°ng ph√°p tƒÉng c∆∞·ªùng d·ªØ li·ªáu
def add_noise(data):
   return data + np.random.normal(0, 0.002, len(data))
def pitch_shifting(data, sr=16000):
   return librosa.effects.pitch_shift(data.astype(np.float32), sr=sr, n_steps=np.random.uniform(-2, 2))
def time_stretching(data, rate=1.5):
   return librosa.effects.time_stretch(data.astype(np.float32), rate=rate)
def volume_change(data, gain=3):
   return data * (10**(gain / 20))
# ‚úÖ Th·ª±c hi·ªán tƒÉng c∆∞·ªùng d·ªØ li·ªáu
print("\nüü¢ B·∫Øt ƒë·∫ßu th·ª±c hi·ªán Data Augmentation...")
augmented_files = []
augmented_targets = []
for index, row in tqdm(dataset.iterrows(), total=len(dataset)):
   file_path = row["files_path"]
   label = row["target"]
   try:
       data, sr = librosa.load(file_path, sr=16000)
       augmentations = [add_noise(data), pitch_shifting(data, sr=sr), time_stretching(data), volume_change(data)]
       for idx, aug_data in enumerate(augmentations):
           aug_filename = f"{row['filename'][:-4]}_aug_{idx}.wav"
           aug_filepath = os.path.join(AUGMENTED_PATH, aug_filename)
           save_augmented_audio(aug_filepath, aug_data, sr)
           augmented_files.append(aug_filepath)
           augmented_targets.append(label)
   except Exception as e:
       print(f"‚ö†Ô∏è L·ªói khi tƒÉng c∆∞·ªùng d·ªØ li·ªáu cho file {file_path}: {e}")
# ‚úÖ T·∫°o DataFrame d·ªØ li·ªáu tƒÉng c∆∞·ªùng v√† h·ª£p nh·∫•t v·ªõi t·∫≠p g·ªëc
dataset_aug = pd.DataFrame({"files_path": augmented_files, "target": augmented_targets})
dataset = pd.concat([dataset, dataset_aug], ignore_index=True)
print(f"\nüü¢ T·ªïng s·ªë m·∫´u sau tƒÉng c∆∞·ªùng: {len(dataset)}")
# ‚úÖ Tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng √¢m thanh
features = ['zero_crossing_rate', 'chroma_stft', 'rmse', 'spectral_centroid', 'spectral_bandwidth',
           'beat_per_minute', 'rolloff'] + [f'mfcc_{i}' for i in range(20)]
for feature in features:
   dataset[feature] = np.zeros(len(dataset))
print("\nüü¢ B·∫Øt ƒë·∫ßu tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng...")
for i in tqdm(range(len(dataset))):
   file_path = dataset['files_path'][i]
   if os.path.exists(file_path):
       y, sr = librosa.load(file_path, mono=True)
       dataset.at[i, 'rmse'] = np.mean(librosa.feature.rms(y=y))
       dataset.at[i, 'chroma_stft'] = np.mean(librosa.feature.chroma_stft(y=y, sr=sr))
       dataset.at[i, 'spectral_centroid'] = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
       dataset.at[i, 'spectral_bandwidth'] = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
       dataset.at[i, 'rolloff'] = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
       dataset.at[i, 'zero_crossing_rate'] = np.mean(librosa.feature.zero_crossing_rate(y))
       tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
       dataset.at[i, 'beat_per_minute'] = tempo
       mfcc = librosa.feature.mfcc(y=y, sr=sr)
       for e in range(20):
           dataset.at[i, f'mfcc_{e}'] = np.mean(mfcc[e])
# ‚úÖ Chia t·∫≠p d·ªØ li·ªáu train/test
train = dataset.sample(frac=0.8, random_state=SEED)
test = dataset.drop(train.index)
X_train = train[features]
X_test = test[features]
y_train = train['target']
y_test = test['target']
scaler = MinMaxScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
# ‚úÖ X√¢y d·ª±ng m√¥ h√¨nh CNN
def build_cnn_lstm(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # ‚úÖ T·∫ßng CNN ƒë·ªÉ tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng kh√¥ng gian
    x = layers.Reshape((input_shape[0], 1))(inputs)
    x = layers.Conv1D(64, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(128, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(256, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)

    # ‚úÖ LSTM ƒë·ªÉ h·ªçc th√¥ng tin chu·ªói
    x = layers.LSTM(128, return_sequences=True, dropout=0.3)(x)
    x = layers.LSTM(64, return_sequences=False, dropout=0.3)(x)

    # ‚úÖ Fully Connected
    x = layers.BatchNormalization()(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003), 
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    
    return model
def build_bilstm(input_shape, num_classes):
   inputs = layers.Input(shape=input_shape)
   # ‚úÖ T·∫ßng CNN ƒë·ªÉ tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng kh√¥ng gian
   x = layers.Reshape((input_shape[0], 1))(inputs)
   x = layers.Conv1D(256, kernel_size=5, padding="same", activation="relu")(x)
   x = layers.MaxPooling1D(pool_size=2)(x)
   x = layers.Conv1D(128, kernel_size=3, padding="same", activation="relu")(x)
   x = layers.MaxPooling1D(pool_size=2)(x)
   # ‚úÖ BiLSTM ƒë·ªÉ h·ªçc th√¥ng tin ng·ªØ c·∫£nh
   x = layers.Bidirectional(layers.LSTM(256, return_sequences=True, dropout=0.3))(x)
   x = layers.Bidirectional(layers.LSTM(128, return_sequences=True, dropout=0.3))(x)
   x = layers.Bidirectional(layers.LSTM(64, return_sequences=False, dropout=0.3))(x)
   # ‚úÖ Th√™m BatchNormalization ƒë·ªÉ ·ªïn ƒë·ªãnh hu·∫•n luy·ªán
   x = layers.BatchNormalization()(x)
   x = layers.Dense(128, activation="relu")(x)
   x = layers.Dropout(0.4)(x)
   outputs = layers.Dense(num_classes, activation="softmax")(x)
   model = models.Model(inputs, outputs)
   model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003), 
                 loss="sparse_categorical_crossentropy",
                 metrics=["accuracy"])
   return model

def build_hst_at(input_shape, num_classes):
   inputs = layers.Input(shape=input_shape)
   # ‚úÖ ƒê·ªãnh h√¨nh d·ªØ li·ªáu ƒë·ªÉ ph√π h·ª£p v·ªõi Transformer
   x = layers.Reshape((input_shape[0], 1))(inputs)  
   # ‚úÖ Positional Encoding
   x = layers.Conv1D(filters=128, kernel_size=3, padding="same", activation="relu")(x)
   x = layers.LayerNormalization()(x)
   # ‚úÖ Multi-Head Self-Attention v·ªõi nhi·ªÅu t·∫ßng
   for _ in range(3):
       attn_output = layers.MultiHeadAttention(num_heads=8, key_dim=64)(x, x)
       x = layers.Add()([x, attn_output])
       x = layers.LayerNormalization()(x)
   # ‚úÖ Global Average Pooling
   x = layers.GlobalAveragePooling1D()(x)
   x = layers.Dense(128, activation="relu")(x)
   x = layers.Dropout(0.3)(x)
   outputs = layers.Dense(num_classes, activation="softmax")(x)
   model = models.Model(inputs, outputs)
   model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
                 loss="sparse_categorical_crossentropy",
                 metrics=["accuracy"])
   return model

# ‚úÖ Hu·∫•n luy·ªán m√¥ h√¨nh
models_dict = {
   "BiLSTM": build_bilstm((X_train.shape[1],), num_classes),
   "CNN + LSTM": build_cnn_lstm((X_train.shape[1],), num_classes),
   "HST-AT": build_hst_at((X_train.shape[1],), num_classes)
}
results = []
for name, model in models_dict.items():
   print(f"\nüîµ Training {name}...")
   history =model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test), verbose=1)
   train_acc = history.history['accuracy'][-1]  # Get the last epoch's training accuracy
   train_loss = history.history['loss'][-1]      # Get the last epoch's training loss
   test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
   results.append({
           "Model": name,
           "Train Accuracy": train_acc,
           "Train Loss": train_loss,
           "Test Accuracy": test_acc,
           "Test Loss": test_loss
       })
# ‚úÖ B·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£
df_results = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)
df_results_styled = df_results.style.background_gradient(cmap="coolwarm")
print(df_results_styled)
sns.barplot(x="Test Accuracy", y="Model", data=df_results, palette="coolwarm")
plt.xlabel("Accuracy (%)")
plt.ylabel("Model")
plt.title("So s√°nh ƒë·ªô ch√≠nh x√°c c·ªßa c√°c m√¥ h√¨nh")
plt.show()
# ‚úÖ S·∫Øp x·∫øp top model theo ƒë·ªô ch√≠nh x√°c
df_summary = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)

# ‚úÖ Hi·ªÉn th·ªã b·∫£ng x·∫øp h·∫°ng top model
print("\nüîπ B·∫£ng X·∫øp H·∫°ng Top Models:")
print(df_summary.to_string(index=False))
# ‚úÖ V·∫Ω bi·ªÉu ƒë·ªì Top Models
plt.figure(figsize=(10, 6))
sns.barplot(x="Test Accuracy", y="Model", data=df_summary, palette="coolwarm")
plt.xlabel("Test Accuracy (%)")
plt.ylabel("Model")
plt.title("Top Model Performance")
plt.show()
# ‚úÖ L·∫•y 1000 m·∫´u ng·∫´u nhi√™n t·ª´ t·∫≠p test g·ªëc
random_indices = np.random.choice(len(X_test), 1000, replace=False)
# ‚úÖ D·ª± ƒëo√°n nh√£n cho t·ª´ng m√¥ h√¨nh
model_results = []
summary_stats = []
for name, model in models_dict.items():
   print(f"\nüü¢ ƒêang d·ª± ƒëo√°n v·ªõi m√¥ h√¨nh: {name}")
   y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
   y_true = y_test.iloc[random_indices].values
   # ‚úÖ Chuy·ªÉn nh√£n v·ªÅ d·∫°ng t√™n th·ª±c t·∫ø
   true_labels = encoder.inverse_transform(y_true)
   predicted_labels = encoder.inverse_transform(y_pred)
   # ‚úÖ T·ªïng h·ª£p k·∫øt qu·∫£ d·ª± ƒëo√°n
   results = []
   for i in range(len(random_indices)):
       results.append({
           "Model": name,
           "File": dataset.iloc[random_indices[i]]["files_path"],
           "True Label": true_labels[i],
           "Predicted Label": predicted_labels[i],
           "Correct": "‚úÖ" if true_labels[i] == predicted_labels[i] else "‚ùå"
       })
   df_results = pd.DataFrame(results)
   # ‚úÖ T√≠nh to√°n s·ªë d·ª± ƒëo√°n ƒë√∫ng/sai
   correct_predictions = sum(y_true == y_pred)
   incorrect_predictions = len(y_true) - correct_predictions
   accuracy = round((correct_predictions / len(y_true)) * 100, 2)
   # ‚úÖ Xu·∫•t b·∫£ng t·ªïng h·ª£p
   summary_stats.append({
       "Model": name,
       "Correct Predictions": correct_predictions,
       "Incorrect Predictions": incorrect_predictions,
       "Accuracy (%)": accuracy
   })
   # ‚úÖ L∆∞u k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
   model_results.append(df_results)
# ‚úÖ Hi·ªÉn th·ªã b·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£ Accuracy
df_summary = pd.DataFrame(summary_stats)
print("\nüîπ B·∫£ng T·ªïng H·ª£p Accuracy:")
print(df_summary.to_string(index=False))
# ‚úÖ Hi·ªÉn th·ªã b·∫£ng chi ti·∫øt k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
for df in model_results:
   model_name = df["Model"][0]
   print(f"\nüîπ B·∫£ng Chi Ti·∫øt D·ª± ƒêo√°n - {model_name}:")
   print(df[['File', 'True Label', 'Predicted Label', 'Correct']].head(10).to_string(index=False))  # Hi·ªÉn th·ªã 10 d√≤ng ƒë·∫ßu ti√™n
# ‚úÖ Hi·ªÉn th·ªã ma tr·∫≠n nh·∫ßm l·∫´n cho t·ª´ng m√¥ h√¨nh
plt.figure(figsize=(15, 10))
for i, (name, model) in enumerate(models_dict.items()):
   y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
   y_true = y_test.iloc[random_indices].values
   plt.subplot(2, 3, i + 1)  # ƒêi·ªÅu ch·ªânh ƒë·ªÉ ph√π h·ª£p v·ªõi s·ªë m√¥ h√¨nh
   sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt="d", cmap="coolwarm")
   plt.xlabel("Predicted Label")
   plt.ylabel("True Label")
   plt.title(f"Confusion Matrix - {name}")
plt.tight_layout()
plt.show()
# ‚úÖ Xu·∫•t b√°o c√°o ph√¢n lo·∫°i cho t·ª´ng m√¥ h√¨nh
for name, model in models_dict.items():
   y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
   y_true = y_test.iloc[random_indices].values
   print(f"\nüîπ Classification Report - {name}:")
   print(classification_report(y_true, y_pred, labels=np.unique(y_true), target_names=encoder.inverse_transform(np.unique(y_true))))
```

```python
# ‚úÖ Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import os
import numpy as np
import pandas as pd
import librosa
import soundfile as sf
import random
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# ‚úÖ C·∫•u h√¨nh seed
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
# ‚úÖ ƒê·ªãnh nghƒ©a ƒë∆∞·ªùng d·∫´n d·ªØ li·ªáu
DATA_PATH = "ESC-50-master/"
CSV_PATH = os.path.join(DATA_PATH, "meta/esc50.csv")
AUDIO_PATH = os.path.join(DATA_PATH, "audio")
AUGMENTED_PATH = "ESC-50-augmented-data"
# ‚úÖ ƒê·ªçc d·ªØ li·ªáu g·ªëc
dataset = pd.read_csv(CSV_PATH)
dataset["files_path"] = dataset["filename"].apply(lambda x: os.path.join(AUDIO_PATH, x))
# ‚úÖ M√£ h√≥a nh√£n
encoder = LabelEncoder()
dataset["target"] = encoder.fit_transform(dataset["category"])
num_classes = len(encoder.classes_)
# ‚úÖ T·∫°o th∆∞ m·ª•c tƒÉng c∆∞·ªùng
os.makedirs(AUGMENTED_PATH, exist_ok=True)
# ‚úÖ H√†m l∆∞u √¢m thanh tƒÉng c∆∞·ªùng
def save_augmented_audio(filename, data, sr=16000):
   sf.write(filename, data, sr)
# ‚úÖ C√°c ph∆∞∆°ng ph√°p tƒÉng c∆∞·ªùng d·ªØ li·ªáu
def add_noise(data):
   return data + np.random.normal(0, 0.002, len(data))
def pitch_shifting(data, sr=16000):
   return librosa.effects.pitch_shift(data.astype(np.float32), sr=sr, n_steps=np.random.uniform(-2, 2))
def time_stretching(data, rate=1.5):
   return librosa.effects.time_stretch(data.astype(np.float32), rate=rate)
def volume_change(data, gain=3):
   return data * (10**(gain / 20))
# ‚úÖ Th·ª±c hi·ªán tƒÉng c∆∞·ªùng d·ªØ li·ªáu
print("\nüü¢ B·∫Øt ƒë·∫ßu th·ª±c hi·ªán Data Augmentation...")
augmented_files = []
augmented_targets = []
for index, row in tqdm(dataset.iterrows(), total=len(dataset)):
   file_path = row["files_path"]
   label = row["target"]
   try:
       data, sr = librosa.load(file_path, sr=16000)
       augmentations = [add_noise(data), pitch_shifting(data, sr=sr), time_stretching(data), volume_change(data)]
       for idx, aug_data in enumerate(augmentations):
           aug_filename = f"{row['filename'][:-4]}_aug_{idx}.wav"
           aug_filepath = os.path.join(AUGMENTED_PATH, aug_filename)
           save_augmented_audio(aug_filepath, aug_data, sr)
           augmented_files.append(aug_filepath)
           augmented_targets.append(label)
   except Exception as e:
       print(f"‚ö†Ô∏è L·ªói khi tƒÉng c∆∞·ªùng d·ªØ li·ªáu cho file {file_path}: {e}")
# ‚úÖ T·∫°o DataFrame d·ªØ li·ªáu tƒÉng c∆∞·ªùng v√† h·ª£p nh·∫•t v·ªõi t·∫≠p g·ªëc
dataset_aug = pd.DataFrame({"files_path": augmented_files, "target": augmented_targets})
dataset = pd.concat([dataset, dataset_aug], ignore_index=True)
print(f"\nüü¢ T·ªïng s·ªë m·∫´u sau tƒÉng c∆∞·ªùng: {len(dataset)}")
# ‚úÖ Tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng √¢m thanh
features = ['zero_crossing_rate', 'chroma_stft', 'rmse', 'spectral_centroid', 'spectral_bandwidth',
           'beat_per_minute', 'rolloff'] + [f'mfcc_{i}' for i in range(20)]
for feature in features:
   dataset[feature] = np.zeros(len(dataset))
print("\nüü¢ B·∫Øt ƒë·∫ßu tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng...")
for i in tqdm(range(len(dataset))):
   file_path = dataset['files_path'][i]
   if os.path.exists(file_path):
       y, sr = librosa.load(file_path, mono=True)
       dataset.at[i, 'rmse'] = np.mean(librosa.feature.rms(y=y))
       dataset.at[i, 'chroma_stft'] = np.mean(librosa.feature.chroma_stft(y=y, sr=sr))
       dataset.at[i, 'spectral_centroid'] = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
       dataset.at[i, 'spectral_bandwidth'] = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
       dataset.at[i, 'rolloff'] = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
       dataset.at[i, 'zero_crossing_rate'] = np.mean(librosa.feature.zero_crossing_rate(y))
       tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
       dataset.at[i, 'beat_per_minute'] = tempo
       mfcc = librosa.feature.mfcc(y=y, sr=sr)
       for e in range(20):
           dataset.at[i, f'mfcc_{e}'] = np.mean(mfcc[e])
# ‚úÖ Chia t·∫≠p d·ªØ li·ªáu train/test
train = dataset.sample(frac=0.8, random_state=SEED)
test = dataset.drop(train.index)
X_train = train[features]
X_test = test[features]
y_train = train['target']
y_test = test['target']
scaler = MinMaxScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
# ‚úÖ X√¢y d·ª±ng m√¥ h√¨nh CNN
def build_cnn_lstm(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # ‚úÖ T·∫ßng CNN ƒë·ªÉ tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng kh√¥ng gian
    x = layers.Reshape((input_shape[0], 1))(inputs)
    x = layers.Conv1D(64, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(128, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(256, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)

    # ‚úÖ LSTM ƒë·ªÉ h·ªçc th√¥ng tin chu·ªói
    x = layers.LSTM(128, return_sequences=True, dropout=0.3)(x)
    x = layers.LSTM(64, return_sequences=False, dropout=0.3)(x)

    # ‚úÖ Fully Connected
    x = layers.BatchNormalization()(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003), 
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    
    return model
def build_bilstm(input_shape, num_classes):
   inputs = layers.Input(shape=input_shape)
   # ‚úÖ T·∫ßng CNN ƒë·ªÉ tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng kh√¥ng gian
   x = layers.Reshape((input_shape[0], 1))(inputs)
   x = layers.Conv1D(256, kernel_size=5, padding="same", activation="relu")(x)
   x = layers.MaxPooling1D(pool_size=2)(x)
   x = layers.Conv1D(128, kernel_size=3, padding="same", activation="relu")(x)
   x = layers.MaxPooling1D(pool_size=2)(x)
   # ‚úÖ BiLSTM ƒë·ªÉ h·ªçc th√¥ng tin ng·ªØ c·∫£nh
   x = layers.Bidirectional(layers.LSTM(256, return_sequences=True, dropout=0.3))(x)
   x = layers.Bidirectional(layers.LSTM(128, return_sequences=True, dropout=0.3))(x)
   x = layers.Bidirectional(layers.LSTM(64, return_sequences=False, dropout=0.3))(x)
   # ‚úÖ Th√™m BatchNormalization ƒë·ªÉ ·ªïn ƒë·ªãnh hu·∫•n luy·ªán
   x = layers.BatchNormalization()(x)
   x = layers.Dense(128, activation="relu")(x)
   x = layers.Dropout(0.4)(x)
   outputs = layers.Dense(num_classes, activation="softmax")(x)
   model = models.Model(inputs, outputs)
   model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003), 
                 loss="sparse_categorical_crossentropy",
                 metrics=["accuracy"])
   return model

def build_cnn(input_shape, num_classes):
    model = models.Sequential([
        layers.Dense(512, activation="relu", input_shape=input_shape),
        layers.BatchNormalization(),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation="softmax")
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model

# ‚úÖ Hu·∫•n luy·ªán m√¥ h√¨nh
models_dict = {
   "BiLSTM": build_bilstm((X_train.shape[1],), num_classes),
   "CNN + LSTM": build_cnn_lstm((X_train.shape[1],), num_classes),
   "CNN": build_cnn((X_train.shape[1],), num_classes),
}
results = []
for name, model in models_dict.items():
   print(f"\nüîµ Training {name}...")
   history =model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test), verbose=1)
   train_acc = history.history['accuracy'][-1]  # Get the last epoch's training accuracy
   train_loss = history.history['loss'][-1]      # Get the last epoch's training loss
   test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
   results.append({
           "Model": name,
           "Train Accuracy": train_acc,
           "Train Loss": train_loss,
           "Test Accuracy": test_acc,
           "Test Loss": test_loss
       })
# ‚úÖ B·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£
df_results = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)
df_results_styled = df_results.style.background_gradient(cmap="coolwarm")
print(df_results_styled)
sns.barplot(x="Test Accuracy", y="Model", data=df_results, palette="coolwarm")
plt.xlabel("Accuracy (%)")
plt.ylabel("Model")
plt.title("So s√°nh ƒë·ªô ch√≠nh x√°c c·ªßa c√°c m√¥ h√¨nh")
plt.show()
# ‚úÖ S·∫Øp x·∫øp top model theo ƒë·ªô ch√≠nh x√°c
df_summary = pd.DataFrame(results).sort_values(by="Test Accuracy", ascending=False)

# ‚úÖ Hi·ªÉn th·ªã b·∫£ng x·∫øp h·∫°ng top model
print("\nüîπ B·∫£ng X·∫øp H·∫°ng Top Models:")
print(df_summary.to_string(index=False))
# ‚úÖ V·∫Ω bi·ªÉu ƒë·ªì Top Models
plt.figure(figsize=(10, 6))
sns.barplot(x="Test Accuracy", y="Model", data=df_summary, palette="coolwarm")
plt.xlabel("Test Accuracy (%)")
plt.ylabel("Model")
plt.title("Top Model Performance")
plt.show()
# ‚úÖ L·∫•y 1000 m·∫´u ng·∫´u nhi√™n t·ª´ t·∫≠p test g·ªëc
random_indices = np.random.choice(len(X_test), 1000, replace=False)
# ‚úÖ D·ª± ƒëo√°n nh√£n cho t·ª´ng m√¥ h√¨nh
model_results = []
summary_stats = []
for name, model in models_dict.items():
   print(f"\nüü¢ ƒêang d·ª± ƒëo√°n v·ªõi m√¥ h√¨nh: {name}")
   y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
   y_true = y_test.iloc[random_indices].values
   # ‚úÖ Chuy·ªÉn nh√£n v·ªÅ d·∫°ng t√™n th·ª±c t·∫ø
   true_labels = encoder.inverse_transform(y_true)
   predicted_labels = encoder.inverse_transform(y_pred)
   # ‚úÖ T·ªïng h·ª£p k·∫øt qu·∫£ d·ª± ƒëo√°n
   results = []
   for i in range(len(random_indices)):
       results.append({
           "Model": name,
           "File": dataset.iloc[random_indices[i]]["files_path"],
           "True Label": true_labels[i],
           "Predicted Label": predicted_labels[i],
           "Correct": "‚úÖ" if true_labels[i] == predicted_labels[i] else "‚ùå"
       })
   df_results = pd.DataFrame(results)
   # ‚úÖ T√≠nh to√°n s·ªë d·ª± ƒëo√°n ƒë√∫ng/sai
   correct_predictions = sum(y_true == y_pred)
   incorrect_predictions = len(y_true) - correct_predictions
   accuracy = round((correct_predictions / len(y_true)) * 100, 2)
   # ‚úÖ Xu·∫•t b·∫£ng t·ªïng h·ª£p
   summary_stats.append({
       "Model": name,
       "Correct Predictions": correct_predictions,
       "Incorrect Predictions": incorrect_predictions,
       "Accuracy (%)": accuracy
   })
   # ‚úÖ L∆∞u k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
   model_results.append(df_results)
# ‚úÖ Hi·ªÉn th·ªã b·∫£ng t·ªïng h·ª£p k·∫øt qu·∫£ Accuracy
df_summary = pd.DataFrame(summary_stats)
print("\nüîπ B·∫£ng T·ªïng H·ª£p Accuracy:")
print(df_summary.to_string(index=False))
# ‚úÖ Hi·ªÉn th·ªã b·∫£ng chi ti·∫øt k·∫øt qu·∫£ d·ª± ƒëo√°n c·ªßa t·ª´ng m√¥ h√¨nh
for df in model_results:
   model_name = df["Model"][0]
   print(f"\nüîπ B·∫£ng Chi Ti·∫øt D·ª± ƒêo√°n - {model_name}:")
   print(df[['File', 'True Label', 'Predicted Label', 'Correct']].head(10).to_string(index=False))  # Hi·ªÉn th·ªã 10 d√≤ng ƒë·∫ßu ti√™n
# ‚úÖ Hi·ªÉn th·ªã ma tr·∫≠n nh·∫ßm l·∫´n cho t·ª´ng m√¥ h√¨nh
plt.figure(figsize=(15, 10))
for i, (name, model) in enumerate(models_dict.items()):
   y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
   y_true = y_test.iloc[random_indices].values
   plt.subplot(2, 3, i + 1)  # ƒêi·ªÅu ch·ªânh ƒë·ªÉ ph√π h·ª£p v·ªõi s·ªë m√¥ h√¨nh
   sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt="d", cmap="coolwarm")
   plt.xlabel("Predicted Label")
   plt.ylabel("True Label")
   plt.title(f"Confusion Matrix - {name}")
plt.tight_layout()
plt.show()
# ‚úÖ Xu·∫•t b√°o c√°o ph√¢n lo·∫°i cho t·ª´ng m√¥ h√¨nh
for name, model in models_dict.items():
   y_pred = np.argmax(model.predict(X_test.iloc[random_indices]), axis=1)
   y_true = y_test.iloc[random_indices].values
   print(f"\nüîπ Classification Report - {name}:")
   print(classification_report(y_true, y_pred, labels=np.unique(y_true), target_names=encoder.inverse_transform(np.unique(y_true))))
```

```python
# ‚úÖ Import th∆∞ vi·ªán c·∫ßn thi·∫øt
import os
import numpy as np
import pandas as pd
import librosa
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer, MinMaxScaler
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ‚úÖ C·∫•u h√¨nh seed
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ‚úÖ ƒê·ªãnh nghƒ©a ƒë∆∞·ªùng d·∫´n d·ªØ li·ªáu FSD50K
DATA_PATH = "fsd50k/"
TRAIN_CSV_PATH = os.path.join(DATA_PATH, "FSD50K.ground_truth", "dev.csv")  # Nh√£n t·∫≠p train
TEST_CSV_PATH = os.path.join(DATA_PATH, "FSD50K.ground_truth", "eval.csv")  # Nh√£n t·∫≠p test
TRAIN_AUDIO_PATH = os.path.join(DATA_PATH, "FSD50K.dev_audio_16k")  # Audio t·∫≠p train
TEST_AUDIO_PATH = os.path.join(DATA_PATH, "FSD50K.eval_audio_16k")  # Audio t·∫≠p test

# ‚úÖ Ki·ªÉm tra d·ªØ li·ªáu c√≥ t·ªìn t·∫°i kh√¥ng
assert os.path.exists(TRAIN_CSV_PATH), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file nh√£n train (dev.csv)!"
assert os.path.exists(TEST_CSV_PATH), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file nh√£n test (eval.csv)!"
assert os.path.exists(TRAIN_AUDIO_PATH), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y th∆∞ m·ª•c audio train (FSD50K.dev_audio_16k)!"
assert os.path.exists(TEST_AUDIO_PATH), "‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y th∆∞ m·ª•c audio test (FSD50K.eval_audio_16k)!"

# ‚úÖ ƒê·ªçc d·ªØ li·ªáu t·ª´ FSD50K
train_df = pd.read_csv(TRAIN_CSV_PATH)
test_df = pd.read_csv(TEST_CSV_PATH)

print("\nüîµ D·ªØ li·ªáu train:")
print(train_df.head())

print("\nüîµ D·ªØ li·ªáu test:")
print(test_df.head())

# ‚úÖ Th√™m ƒë∆∞·ªùng d·∫´n file audio
train_df["files_path"] = train_df["fname"].apply(lambda x: os.path.join(TRAIN_AUDIO_PATH, f"{x}.wav"))
test_df["files_path"] = test_df["fname"].apply(lambda x: os.path.join(TEST_AUDIO_PATH, f"{x}.wav"))

# ‚úÖ M√£ h√≥a nh√£n (multi-label)
mlb = MultiLabelBinarizer()
train_df["labels"] = train_df["labels"].apply(lambda x: x.split(","))  # Chuy·ªÉn nh√£n th√†nh danh s√°ch
test_df["labels"] = test_df["labels"].apply(lambda x: x.split(","))

y_train = mlb.fit_transform(train_df["labels"])  # One-hot encode t·∫≠p train
y_test = mlb.transform(test_df["labels"])  # One-hot encode t·∫≠p test
# ‚úÖ Tr√≠ch xu·∫•t c·∫£ Mel-Spectrogram v√† Feature truy·ªÅn th·ªëng
def extract_features(file_path, sr=16000, n_mels=64):
    try:
        y, sr = librosa.load(file_path, sr=sr)
        
        # üîπ Mel-Spectrogram
        mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels)
        log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)
        mel_features = np.mean(log_mel_spec, axis=1)  # L·∫•y trung b√¨nh theo tr·ª•c th·ªùi gian
        
        # üîπ Feature truy·ªÅn th·ªëng
        rmse = np.mean(librosa.feature.rms(y=y))
        chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr))
        spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
        spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
        rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
        
        # üîπ K·∫øt h·ª£p t·∫•t c·∫£ ƒë·∫∑c tr∆∞ng
        combined_features = np.hstack([mel_features, rmse, chroma, spectral_centroid, spectral_bandwidth, rolloff])
        return combined_features
    
    except Exception as e:
        print(f"‚ö†Ô∏è L·ªói khi tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng t·ª´ {file_path}: {e}")
        return np.zeros(n_mels + 5)  # S·ªë feature = n_mels (Mel) + 5 (Feature truy·ªÅn th·ªëng)

# ‚úÖ Tr√≠ch xu·∫•t ƒë·∫∑c tr∆∞ng cho t·∫≠p train v√† test
num_features = n_mels + 5  # T·ªïng s·ªë feature (Mel + feature truy·ªÅn th·ªëng)
X_train = np.array([extract_features(file) for file in tqdm(train_df["files_path"], desc="Extracting Features Train")])
X_test = np.array([extract_features(file) for file in tqdm(test_df["files_path"], desc="Extracting Features Test")])

# ‚úÖ Chu·∫©n h√≥a d·ªØ li·ªáu
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("\n‚úÖ D·ªØ li·ªáu ƒë√£ s·∫µn s√†ng ƒë·ªÉ hu·∫•n luy·ªán!")
```

```python
def build_cnn_lstm(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # üîπ CNN Layers
    x = layers.Reshape((input_shape[0], 1))(inputs)
    x = layers.Conv1D(64, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(128, kernel_size=3, activation="relu", padding="same")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)

    # üîπ LSTM Layers
    x = layers.LSTM(128, return_sequences=True, dropout=0.3)(x)
    x = layers.LSTM(64, return_sequences=False, dropout=0.3)(x)

    # üîπ Fully Connected
    x = layers.BatchNormalization()(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation="sigmoid")(x)  # üîπ S·ª≠ d·ª•ng sigmoid cho multi-label

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003), 
                  loss="binary_crossentropy",  # üîπ Loss cho multi-label
                  metrics=["accuracy"])
    
    return model
```

```python
def build_bilstm(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # üîπ CNN Layers
    x = layers.Reshape((input_shape[0], 1))(inputs)
    x = layers.Conv1D(256, kernel_size=5, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(128, kernel_size=3, padding="same", activation="relu")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)

    # üîπ BiLSTM Layers
    x = layers.Bidirectional(layers.LSTM(256, return_sequences=True, dropout=0.3))(x)
    x = layers.Bidirectional(layers.LSTM(128, return_sequences=True, dropout=0.3))(x)
    x = layers.Bidirectional(layers.LSTM(64, return_sequences=False, dropout=0.3))(x)

    # üîπ Fully Connected
    x = layers.BatchNormalization()(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(num_classes, activation="sigmoid")(x)

    model = models.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003),
                  loss="binary_crossentropy",
                  metrics=["accuracy"])
    
    return model
```
